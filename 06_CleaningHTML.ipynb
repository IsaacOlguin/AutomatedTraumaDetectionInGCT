```python
html_files = [f for f in listdir(PATH_HTML_FILES) if isfile(join(PATH_HTML_FILES, f))]
num_files = len(html_files)
print(f"Num of files is <<{num_files}>>")
```

```python
def readFile(path, fileEncoding):
  content = None
  try:
      f = open(path, "r", encoding=fileEncoding)
      content = f.read()
  except Exception as ex:
    print(f"An exception occurs while reading the file")
    content = None
  return content
```

```python
for index, filename in enumerate(html_files):
  print(f"Processing file ({(index+1)}/{num_files}) [{filename}]... ", end="")

  print("Reading... ", end="")
  contentFile = readFile(PATH_HTML_FILES + filename, FILE_ENCODING)
```

```python
print(contentFile)
```

```python
contentFileAux = re.sub(r'<HEAD>[A-Za-z\W0-9]*</HEAD>', "", contentFile)
print(contentFileAux)
```

```python
contentFileAux = re.sub(r'<CENTER><P><B>( )*</B></P></CENTER>', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
contentFileAux = re.sub(r'<CENTER><P><B>( )*Page (\d)+( )*</B></P></CENTER>', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
contentFileAux = re.sub(r'(<P>( \d+|\d+)( )*Blank page inserted to ensure pagination corresponds between the French and( )*</P>)|(<P>( \d+|\d+)( )*English transcripts.( )*</P>)', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
contentFileAux = re.sub(r'<P>( \d+|\d+)( )*</P>', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
contentFileAux = re.sub(r'</P>( )*', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
contentFileAux = re.sub(r'<P>( \d+|\d+)( )*', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
"""
<HTML>

<BODY bgcolor="#FFFFFF">
<center>
<img SRC="../image/title.gif" HEIGHT="74" WIDTH="502" ALT="Tribunal Criminal Tribunal for the Former Yugoslavia">
</center>

<FONT FACE="Courier New" SIZE=2>


</FONT></BODY>
</HTML>

"""

contentFileAux = re.sub(r'<HTML>|</HTML>|<center>|</center>|</FONT>|</BODY>|<FONT FACE(.*)>', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
"""
<BODY bgcolor="#FFFFFF">

<img SRC="../image/title.gif" HEIGHT="74" WIDTH="502" ALT="Tribunal Criminal Tribunal for the Former Yugoslavia">


"""

contentFileAux = re.sub(r'<BODY(.*)>|<img(.*)>', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```

```python
contentFileAux = re.sub(r'^( )(\\n)*', "", contentFileAux)
print(len(contentFileAux))
print(contentFileAux)
```
